```python
import pandas as pd
import os
import glob
import re
import numpy as np
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
chrome_options = Options()

path = os.getcwd()
```

```python
df = pd.read_csv('CID_2118_compound.csv', nrows=(3620-1811))
```

```python
df2 = pd.read_csv('CID_3016_compound.csv',nrows=(5606-2806))
```

```python
df = df[df['annothits'].str.contains("Biological")]
```

```python
df2 = df2[df2['annothits'].str.contains("Biological")]
```

```python
combd = pd.concat([df,df2])
```

```python
combd
```

```python
group = combd.groupby('cid').count()[['cmpdname']].sort_values('cmpdname', ascending = False)
group[['cmpdname']] = group[['cmpdname']].astype(int)
group =group.reset_index()
```

```python
dup = list(group.loc[group['cmpdname'] > 1, 'cid'])
```

```python
len(dup)
```

```python
x = list(df['cid'])
```

```python
len(x)
```

```python
x = [i for i in x if i not in dup]
```

```python
len(x)
```

```python
chromeOptions = Options()
chromeOptions.add_experimental_option('prefs',{'download.default_directory':'/Users/layalhammad/git_proj/project_1/project-1/scrapped2'})

driver = webdriver.Chrome("/usr/local/bin/chromedriver", chrome_options = chromeOptions)
  
    
for i in x:    
    # Open URL
    driver.get(
        f'https://pubchem.ncbi.nlm.nih.gov/compound/{i}#section=BioAssay-Results&fullscreen=true')

    time.sleep(2)

    try:
        download_button = driver.find_element_by_xpath("//button[@id='Download']")


        download_button.click()

        time.sleep(2)

        download_csv_button = driver.find_element_by_xpath("//a[@data-label='Download Section: BioAssay-Results; CSV - Save']")

        download_csv_button.click()

        time.sleep(2)
    
    except:
        continue

driver.close()

  
```
